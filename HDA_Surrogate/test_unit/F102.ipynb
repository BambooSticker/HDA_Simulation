```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pyomo.environ import *
from idaes.core.util.misc import extract_data
import pyomo.environ as pyo
from pyomo.util.calc_var_value import calculate_variable_from_constraint
from State_Block import Create_State_Block
from idaes.core.util.model_statistics import degrees_of_freedom
from Unit_Model import F102_creater
from Method import Initialize
```

```python
# Create a concrete model
m = ConcreteModel()
# Define a set for components
m.Phase = Set(initialize = ['Liq','Vap'])
m.Comp = Set(initialize = ['benzene','toluene','methane','hydrogen'])
```

```python
m.F102 = F102_creater(m)
```

```python
# 各组分进料
m.F102.inlet.flow['Liq','benzene'].fix(0.20460)
m.F102.inlet.flow['Liq','toluene'].fix(0.062520)
m.F102.inlet.flow['Liq','methane'].fix(2.0647e-07)
m.F102.inlet.flow['Liq','hydrogen'].fix(2.0647e-07)
m.F102.inlet.flow['Vap','benzene'].fix(1.0000e-08)
m.F102.inlet.flow['Vap','toluene'].fix(1.0000e-08)
m.F102.inlet.flow['Vap','methane'].fix(1.0000e-08)
m.F102.inlet.flow['Vap','hydrogen'].fix(1.0000e-08)

# 温度压力
m.F102.inlet.P.fix(350000)
m.F102.inlet.T.fix(325)

m.F102.outlet.P.fix(150000)
m.F102.outlet.T.fix(375)
```

```python
Initialize(m.F102)
```

```python
# 模型求解
solver = pyo.SolverFactory('ipopt')
solver.options['tol'] = 1e-3
results = solver.solve(m)
results = solver.solve(m,tee=True)
```
